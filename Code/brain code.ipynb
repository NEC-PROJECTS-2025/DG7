**Grayscale**

```python
from PIL import Image
import os

# Path to your dataset directory
dataset_dir = '/content/drive/MyDrive/brain_dataset/Training/glioma_tumor'

# Create a directory to store grayscale images
output_dir = '/content/drive/MyDrive/brain_dataset/Training/glioma_tumor_grayscale'
os.makedirs(output_dir, exist_ok=True)

# Iterate through each image in the dataset directory
for filename in os.listdir(dataset_dir):
    if filename.endswith('.jpg') or filename.endswith('.png'):  # Adjust extensions as necessary
        # Open the image
        img_path = os.path.join(dataset_dir, filename)
        img = Image.open(img_path)

        # Convert to grayscale
        grayscale_img = img.convert('L')

        # Resize to 224x224
        resized_img = grayscale_img.resize((224, 224))

        # Save the grayscale resized image
        output_path = os.path.join(output_dir, filename)
        resized_img.save(output_path)

        print(f'Processed {filename}')

print('Conversion complete.')
```

**Gaussian Blur**

```python
import cv2
import os
import numpy as np  # Import numpy

# Directory containing the images
input_dir = "/content/drive/MyDrive/brain_dataset/Training/glioma_tumor"
output_dir = "/content/drive/MyDrive/brain_dataset/Processed/Training/glioma_tumor"

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Function to apply high-pass filter
def high_pass_filter(img):
    # Create a kernel for the high-pass filter
    kernel = [[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]]
    kernel = np.array(kernel)

    # Apply the high-pass filter
    high_pass = cv2.filter2D(img, -1, kernel)
    return high_pass

# Loop through all files in the directory
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png"):  # Modify as needed for your file types
        # Read the image
        img_path = os.path.join(input_dir, filename)
        img = cv2.imread(img_path)

        # Convert to grayscale
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Resize the image
        resized_img = cv2.resize(gray_img, (224, 224))

        # Apply Gaussian Blur
        blurred_img = cv2.GaussianBlur(resized_img, (5, 5), 0)

        # Apply High-Pass Filter
        high_pass_img = high_pass_filter(blurred_img)

        # Save the processed image
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, high_pass_img)

        print(f"Processed {filename}")
```

**Erosion and Dilation**

```python
import cv2
import os
import numpy as np

# Directory containing the images
input_dir = "/content/drive/MyDrive/brain_dataset/Training/glioma_tumor"
output_dir = "/content/drive/MyDrive/brain_dataset/Processed/Training/glioma_tumor"

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Function to apply erosion and dilation
def apply_morphological_operations(img):
    # Define a kernel for morphological operations
    kernel = np.ones((5, 5), np.uint8)

    # Apply erosion
    eroded_img = cv2.erode(img, kernel, iterations=1)

    # Apply dilation
    dilated_img = cv2.dilate(eroded_img, kernel, iterations=1)

    return dilated_img

# Get list of files in the input directory
input_files = [f for f in os.listdir(input_dir) if f.endswith(".jpg") or f.endswith(".png")]
total_files = len(input_files)

# Loop through all files in the directory
processed_files = 0
for filename in input_files:
    try:
        # Read the image
        img_path = os.path.join(input_dir, filename)
        img = cv2.imread(img_path)

        # Convert to grayscale
        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Resize the image
        resized_img = cv2.resize(gray_img, (224, 224))

        # Apply erosion and dilation
        morph_img = apply_morphological_operations(resized_img)

        # Save the processed image
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, morph_img)

        processed_files += 1
        print(f"Processed {filename} ({processed_files}/{total_files})")
    except Exception as e:
        print(f"Error processing {filename}: {e}")

# Check if all files are processed
if processed_files == total_files:
    print("All files have been successfully processed.")
else:
    print(f"Some files may not have been processed. Processed {processed_files} out of {total_files} files. Please check the logs for details.")
```

**Contour Detection**

```python
import cv2
import os
import numpy as np

# Directory containing the images
input_dir = "/content/drive/MyDrive/brain_dataset/Training/glioma_tumor"
output_dir = "/content/drive/MyDrive/brain_dataset/Processed/Training/glioma_tumor"

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

# Function to perform contour detection and remove black portions
def remove_black_portions(img):
    # Convert image to grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Thresholding to create a binary image
    _, binary_img = cv2.threshold(gray_img, 1, 255, cv2.THRESH_BINARY)

    # Find contours
    contours, _ = cv2.findContours(binary_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Create an empty mask
    mask = np.zeros_like(gray_img)

    # Draw all contours on the mask
    cv2.drawContours(mask, contours, -1, (255), thickness=cv2.FILLED)

    # Bitwise AND to keep only the contours
    result_img = cv2.bitwise_and(img, img, mask=mask)

    return result_img

# Get list of files in the input directory
input_files = [f for f in os.listdir(input_dir) if f.endswith(".jpg") or f.endswith(".png")]
total_files = len(input_files)

# Loop through all files in the directory
processed_files = 0
for filename in input_files:
    try:
        # Read the image
        img_path = os.path.join(input_dir, filename)
        img = cv2.imread(img_path)

        # Apply contour detection and remove black portions
        result_img = remove_black_portions(img)

        # Save the processed image
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, result_img)

        processed_files += 1
        print(f"Processed {filename} ({processed_files}/{total_files})")
    except Exception as e:
        print(f"Error processing {filename}: {e}")

# Check if all files are processed
if processed_files == total_files:
    print("All files have been successfully processed.")
else:
    print(f"Some files may not have been processed. Processed {processed_files} out of {total_files} files. Please check the logs for details.")
```

```python
import cv2
import os
import numpy as np
from PIL import Image, ImageOps
import random

# Directory containing the images
input_dir = "/content/drive/MyDrive/brain_dataset/Training/glioma_tumor" # Replace with your actual directory
output_dir = "/content/drive/MyDrive/brain_dataset/Augmented/Training/glioma_tumor" # Replace with your desired output directory

# Ensure output directory exists
os.makedirs(output_dir, exist_ok=True)

def augment_image(img):
    # Convert image to PIL format for transformations
    pil_img = Image.fromarray(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    # Random Rotation
    angle = random.uniform(-30, 30)  # Random angle between -30 and 30 degrees
    pil_img = pil_img.rotate(angle, expand=True)

    # Horizontal Flip
    if random.random() > 0.5:
        pil_img = ImageOps.mirror(pil_img)

    # Vertical Flip
    if random.random() > 0.5:
        pil_img = ImageOps.flip(pil_img)

    # Transpose
    if random.random() > 0.5:
        pil_img = pil_img.transpose(Image.Transpose.ROTATE_90)

    # Convert back to OpenCV format
    img = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)

    return img

# Get list of files in the input directory
input_files = [f for f in os.listdir(input_dir) if f.endswith(".jpg") or f.endswith(".png")]
total_files = len(input_files)

# Loop through all files in the directory
processed_files = 0
for filename in input_files:
    try:
        # Read the image
        img_path = os.path.join(input_dir, filename)
        img = cv2.imread(img_path)

        # Apply transformations
        augmented_img = augment_image(img)

        # Save the processed image
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, augmented_img)

        processed_files += 1
        print(f"Processed {filename} ({processed_files}/{total_files})")
    except Exception as e:
        print(f"Error processing {filename}: {e}")

# Check if all files are processed
if processed_files == total_files:
    print("All files have been successfully processed.")
else:
    print(f"Some files may not have been processed. Processed {processed_files} out of {total_files} files. Please check the logs for details.")
```

```python
import os
import shutil
import random

def split_dataset(dataset_dir, train_dir, val_dir, val_split=0.2):
  """Splits a dataset into training and validation sets.

  Args:
    dataset_dir: Path to the dataset directory.
    train_dir: Path to the output training directory.
    val_dir: Path to the output validation directory.
    val_split: Fraction of the dataset to use for validation.
  """

  os.makedirs(train_dir, exist_ok=True)
  os.makedirs(val_dir, exist_ok=True)

  for class_folder in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_folder)
    if os.path.isdir(class_path):
      images = os.listdir(class_path)
      num_val = int(len(images) * val_split)
      val_images = random.sample(images, num_val)

      for image in images:
        source_path = os.path.join(class_path, image)
        if image in val_images:
          dest_path = os.path.join(val_dir, class_folder, image) # Destination path is in the validation directory
        else:
          dest_path = os.path.join(train_dir, class_folder, image) # Destination path is in the training directory
        os.makedirs(os.path.dirname(dest_path), exist_ok=True)
        shutil.copy(source_path, dest_path)

# Example usage:
dataset_dir = '/content/drive/MyDrive/brain_dataset/Training' # Replace with your dataset directory
train_dir = '/content/drive/MyDrive/brain_dataset/Split_Data/Training' # Replace with your desired training directory
val_dir = '/content/drive/MyDrive/brain_dataset/Split_Data/Validation' # Replace with your desired validation directory

split_dataset(dataset_dir, train_dir, val_dir, val_split=0.2)
```

```python
# prompt: i want training data with 2400 images and validation data has 600 images give me code

import os
import random
import shutil

def split_dataset(dataset_dir, train_dir, val_dir, train_count=2400, val_count=600):
  """Splits a dataset into training and validation sets with specific counts.

  Args:
    dataset_dir: Path to the dataset directory.
    train_dir: Path to the output training directory.
    val_dir: Path to the output validation directory.
    train_count: Number of images for the training set.
    val_count: Number of images for the validation set.
  """

  os.makedirs(train_dir, exist_ok=True)
  os.makedirs(val_dir, exist_ok=True)

  for class_folder in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_folder)
    if os.path.isdir(class_path):
      images = os.listdir(class_path)
      random.shuffle(images)  # Shuffle images randomly

      train_images = images[:train_count // 9]  # Distribute images evenly across classes
      val_images = images[train_count // 9: train_count // 9 + val_count // 9]

      for image in images:
        source_path = os.path.join(class_path, image)
        if image in val_images:
          dest_path = os.path.join(val_dir, class_folder, image)
        elif image in train_images:
          dest_path = os.path.join(train_dir, class_folder, image)
        else:
          continue  # Skip images exceeding the desired count
        os.makedirs(os.path.dirname(dest_path), exist_ok=True)
        shutil.copy(source_path, dest_path)

# Example usage:
dataset_dir = '/content/drive/MyDrive/brain_dataset/Training'  # Replace with your dataset directory
train_dir = '/content/drive/MyDrive/brain_dataset/Split_Data/Training'  # Replace with your desired training directory
val_dir = '/content/drive/MyDrive/brain_dataset/Split_Data/Validation'  # Replace with your desired validation directory

split_dataset(dataset_dir, train_dir, val_dir, train_count=2400, val_count=600)
```

```python
import tensorflow as tf
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau
from tensorflow.keras.regularizers import l2

# Set parameters
input_shape = (224, 224, 3)
num_classes = 4  # As per your dataset
batch_size = 32
epochs = 50  # Increased epochs
initial_learning_rate = 1e-4
l2_reg = 1e-4  # L2 regularization factor
fine_tune_at = 249  # Fine-tune from this layer (for InceptionV3, layer 249 is a good place)

# Load the InceptionV3 model (GoogLeNet) without the top layers
base_model = InceptionV3(weights='imagenet', include_top=False, input_shape=input_shape)

# Add custom top layers with L2 regularization
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)  # Dropout for regularization
x = Dense(1024, activation='relu', kernel_regularizer=l2(l2_reg))(x)  # L2 regularization
x = Dropout(0.5)(x)  # Dropout for regularization
predictions = Dense(num_classes, activation='softmax', kernel_regularizer=l2(l2_reg))(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

# Freeze all the layers before `fine_tune_at` (early layers of InceptionV3)
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

# Compile the model
model.compile(optimizer=Adam(learning_rate=initial_learning_rate),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Data generators with more aggressive augmentation
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=30,  # More augmentation
                                   width_shift_range=0.2,
                                   height_shift_range=0.2,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   fill_mode='nearest')

val_datagen = ImageDataGenerator(rescale=1./255)

# Load data from directories
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/brain_dataset/Split_Data/Training',
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    '/content/drive/MyDrive/brain_dataset/Split_Data/Validation',
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical'
)

# Learning rate reduction on plateau
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)

# Train the model
history = model.fit(
    train_generator,
    epochs=epochs,
    validation_data=val_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    validation_steps=val_generator.samples // batch_size,
    callbacks=[reduce_lr]
)

# Save the model
model.save('googlenet_inceptionv3_finetuned.keras')

# Plot the training history
import matplotlib.pyplot as plt

# Accuracy plot
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Loss plot
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# Evaluate the model on validation set
val_loss, val_accuracy = model.evaluate(val_generator)
print(f'Validation Accuracy: {val_accuracy}')
print(f'Validation Loss: {val_loss}')
```

```python
import tensorflow as tf
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, Dense
from tensorflow.keras.regularizers import l2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# Normalize images and apply augmentations
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
val_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Load training and validation datasets
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/brain_dataset/Training',  # Replace with the actual path
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

validation_generator = val_datagen.flow_from_directory(
    '/content/drive/MyDrive/brain_dataset/Validation',  # Replace with the actual path
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical'
)

# Load DenseNet121 base model
base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze base model

# Add custom layers
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(1024, activation='relu', kernel_regularizer=l2(0.01)),
    Dropout(0.5),
    Dense(4, activation='softmax', kernel_regularizer=l2(0.01))  # Replace 4 with the number of your classes
])

# Compile the model for initial training
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train the model with frozen base model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10
)

# Unfreeze the top 20 layers of the base model for fine-tuning
for layer in base_model.layers[-20:]:
    layer.trainable = True

# Re-compile the model with a smaller learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Use a learning rate scheduler
lr_scheduler = tf.keras.callbacks.ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.1,
    patience=3,
    verbose=1
)

# Fine-tune the model
history_fine_tune = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=20,
    callbacks=[lr_scheduler]
)

# Save the trained model
model.save('best_model.keras')

# Plot training and validation accuracy
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy (Frozen)')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy (Frozen)')
plt.plot(history_fine_tune.history['accuracy'], label='Training Accuracy (Fine-Tuned)')
plt.plot(history_fine_tune.history['val_accuracy'], label='Validation Accuracy (Fine-Tuned)')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss (Frozen)')
plt.plot(history.history['val_loss'], label='Validation Loss (Frozen)')
plt.plot(history_fine_tune.history['loss'], label='Training Loss (Fine-Tuned)')
plt.plot(history_fine_tune.history['val_loss'], label='Validation Loss (Fine-Tuned)')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.applications import DenseNet201
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dropout, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Data Augmentation for training and validation datasets
train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    shear_range=0.2,
    zoom_range=0.2,
    rotation_range=30,
    horizontal_flip=True,
    brightness_range=[0.8, 1.2],
    fill_mode='nearest'
)

val_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Load datasets
train_generator = train_datagen.flow_from_directory(
    '/content/drive/MyDrive/brain_dataset/Training',
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical'
)

validation_generator = val_datagen.flow_from_directory(
    '/content/drive/MyDrive/brain_dataset/Validation',
    target_size=(224, 224),
    batch_size=16,
    class_mode='categorical'
)

# Load DenseNet201 base model
base_model = DenseNet201(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model.trainable = False  # Freeze base model

# Add custom layers
model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dropout(0.5),
    Dense(1024, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    Dropout(0.5),
    Dense(4, activation='softmax', kernel_regularizer=tf.keras.regularizers.l2(0.01))  # Replace 4 with the number of classes
])

# Compile the model with an initial learning rate
model.compile(optimizer=Adam(learning_rate=1e-3),
              loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
              metrics=['accuracy'])

# Callbacks for learning rate adjustment and early stopping
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, verbose=1, min_lr=1e-6)
early_stopping = EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True, verbose=1)

# Train the model with frozen base model
history = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=10,
    callbacks=[reduce_lr, early_stopping]
)

# Unfreeze the base model for fine-tuning
base_model.trainable = True

# Re-compile the model with a smaller learning rate
model.compile(optimizer=Adam(learning_rate=1e-5),
              loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),
              metrics=['accuracy'])

# Fine-tune the model
history_fine_tune = model.fit(
    train_generator,
    validation_data=validation_generator,
    epochs=20,
    callbacks=[reduce_lr, early_stopping]
)

# Save the model
model.save('densenet201_fine_tuned_model.keras')

# Plot training and validation accuracy
plt.figure(figsize=(12, 6))
plt.plot(history.history['accuracy'], label='Training Accuracy (Frozen)')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy (Frozen)')
plt.plot(history_fine_tune.history['accuracy'], label='Training Accuracy (Fine-Tuned)')
plt.plot(history_fine_tune.history['val_accuracy'], label='Validation Accuracy (Fine-Tuned)')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot training and validation loss
plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Training Loss (Frozen)')
plt.plot(history.history['val_loss'], label='Validation Loss (Frozen)')
plt.plot(history_fine_tune.history['loss'], label='Training Loss (Fine-Tuned)')
plt.plot(history_fine_tune.history['val_loss'], label='Validation Loss (Fine-Tuned)')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
